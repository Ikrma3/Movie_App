```python
import numpy as np
import pandas as pd
from pathlib import Path


df_path = Path("data_emotions/merged_emotions.csv")
assert df_path.exists(), "merged_emotions.csv not found. Run Step 2."
df = pd.read_csv(df_path)


classes = sorted(df["label_raw"].astype(str).str.lower().str.strip().unique())
label2id = {c: i for i, c in enumerate(classes)}
id2label = {i: c for c, i in label2id.items()}
df["label"] = df["label_raw"].map(lambda x: label2id[str(x).lower().strip()])
num_labels = len(classes)

print("Classes:", classes)
print("Num labels:", num_labels)


from sklearn.model_selection import train_test_split
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["label"])
val_df, test_df   = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df["label"])
print(f"Shapes -> train: {train_df.shape}, val: {val_df.shape}, test: {test_df.shape}")


try:
    from transformers import AutoTokenizer
    from datasets import Dataset, DatasetDict
except Exception:
    !pip -q install transformers==4.44.2 datasets==3.0.1 accelerate==0.34.2 -U
    from transformers import AutoTokenizer
    from datasets import Dataset, DatasetDict

MODEL_NAME = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
MAX_LEN = 128

def tokenize_batch(examples):
    return tokenizer(
        examples["text"],
        padding="max_length",
        truncation=True,
        max_length=MAX_LEN
    )


train_ds = Dataset.from_pandas(train_df[["text","label"]], preserve_index=False)
val_ds   = Dataset.from_pandas(val_df[["text","label"]],   preserve_index=False)
test_ds  = Dataset.from_pandas(test_df[["text","label"]],  preserve_index=False)

raw = DatasetDict({"train": train_ds, "validation": val_ds, "test": test_ds})
tokenized = raw.map(tokenize_batch, batched=True, desc="Tokenizing")
tokenized = tokenized.remove_columns(["text"]).rename_column("label", "labels")
tokenized.set_format("torch")

print("\nTokenized dataset ready.")
for split in ["train", "validation", "test"]:
    print(split, "->", tokenized[split])


STATE = {
    "MODEL_NAME": MODEL_NAME,
    "MAX_LEN": MAX_LEN,
    "num_labels": num_labels,
    "id2label": {int(k): v for k, v in id2label.items()},
    "label2id": {k: int(v) for k, v in label2id.items()}
}
print("\nSTATE prepared.")
```

```python

import numpy as np
import torch
from sklearn.metrics import accuracy_score, f1_score
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

assert "STATE" in globals(), "STATE missing. Run Step 3."
assert "tokenized" in globals(), "Tokenized datasets missing. Run Step 3."

MODEL_NAME = STATE["MODEL_NAME"]
num_labels = STATE["num_labels"]
id2label   = {int(k): v for k, v in STATE["id2label"].items()}

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": float(accuracy_score(labels, preds)),
        "f1_macro": float(f1_score(labels, preds, average="macro")),
    }

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=num_labels,
    id2label={i: id2label[i] for i in range(num_labels)},
    label2id={id2label[i]: i for i in range(num_labels)}
)

batch_size = 16 if torch.cuda.is_available() else 8
args = TrainingArguments(
    output_dir="outputs_bert_emotions",
    learning_rate=2e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=3,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="f1_macro",
    report_to="none",
    logging_steps=50
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized["train"],
    eval_dataset=tokenized["validation"],
    compute_metrics=compute_metrics,
)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device, "| Batch size:", batch_size)

train_result = trainer.train()
print("\nBest checkpoint:", trainer.state.best_model_checkpoint)
```

```python
import json, ast
from pathlib import Path
import pandas as pd
import numpy as np

DATA_DIR = Path("data_recipes")
assert DATA_DIR.exists(), "data_recipes/ not found. Run Step 1."

def to_list(x):
    """Convert ingredients/steps to a clean list of strings."""
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return []
    if isinstance(x, list):
        return [str(i).strip() for i in x if str(i).strip()]
    s = str(x).strip()
    if not s:
        return []

    for loader in (json.loads, ast.literal_eval):
        try:
            v = loader(s)
            if isinstance(v, list):
                return [str(i).strip() for i in v if str(i).strip()]
        except Exception:
            pass
    # split by common delimiters
    if "\n" in s:
        parts = [p.strip("-•* \t") for p in s.split("\n")]
    elif "|" in s:
        parts = [p.strip() for p in s.split("|")]
    elif ";" in s:
        parts = [p.strip() for p in s.split(";")]
    elif "," in s:
        parts = [p.strip() for p in s.split(",")]
    else:
        parts = [s]
    return [p for p in parts if p]

def first_nonempty(*vals):
    for v in vals:
        if v is None:
            continue
        if isinstance(v, str) and v.strip():
            return v.strip()
        if not isinstance(v, str) and pd.notna(v):
            return str(v).strip()
    return ""

def pick_col(df, names, default=None):
    for n in names:
        if n in df.columns:
            return n
    return default

def normalize_df(df):
    cols = [c.lower().strip() for c in df.columns]
    df = df.rename(columns={old:new for old,new in zip(df.columns, cols)})

    title_col = pick_col(df, ["title","name","dish","recipe_name"])
    ing_col   = pick_col(df, ["ingredients","ingredient","ings","ing_list"])
    steps_col = pick_col(df, ["instructions","steps","directions","method","preparation","procedure","recipe"])

    title = df[title_col] if title_col in df.columns else ""
    ings  = df[ing_col]   if ing_col   in df.columns else ""
    steps = df[steps_col] if steps_col in df.columns else ""

    out = pd.DataFrame({
        "title": [first_nonempty(t) for t in title] if isinstance(title, pd.Series) else ["" for _ in range(len(df))],
        "ingredients": [to_list(i) for i in ings] if isinstance(ings, pd.Series) else [[] for _ in range(len(df))],
        "instructions": [to_list(s) for s in steps] if isinstance(steps, pd.Series) else [[] for _ in range(len(df))]
    })


    def render_row(r):
        t = r["title"] or "Unknown Dish"
        ing_lines = "\n".join(f"- {i}" for i in r["ingredients"]) if r["ingredients"] else "- (not provided)"
        if r["instructions"]:
            step_lines = "\n".join(f"{i+1}. {s}" for i,s in enumerate(r["instructions"]))
        else:
            # leave blank target; model will learn to generate
            step_lines = ""
        return (
            "<|recipe|>\n"
            f"Title: {t}\n"
            "Ingredients:\n"
            f"{ing_lines}\n"
            "Instructions:\n"
            f"{step_lines}\n"
            "<|endofrecipe|>"
        )

    out["text"] = out.apply(render_row, axis=1)
    # Filter empties
    out = out[(out["title"].str.len()>0) | (out["ingredients"].str.len()>0) | (out["instructions"].str.len()>0)]
    return out.reset_index(drop=True)


all_rows = []
for p in sorted(list(DATA_DIR.glob("*.csv")) + list(DATA_DIR.glob("*.json"))):
    try:
        if p.suffix.lower()==".csv":
            df = pd.read_csv(p)
        else:

            raw = json.loads(p.read_text())
            if isinstance(raw, list):
                df = pd.DataFrame(raw)
            elif isinstance(raw, dict):

                key = next((k for k,v in raw.items() if isinstance(v, list)), None)
                df = pd.DataFrame(raw[key]) if key else pd.DataFrame([raw])
            else:
                continue
        norm = normalize_df(df)
        norm["__source__"] = p.name
        all_rows.append(norm)
        print(f"Loaded {p.name}: {len(norm)} rows")
    except Exception as e:
        print(f"Skipping {p.name} due to error: {e}")

assert all_rows, "No compatible CSV/JSON files found to build dataset."
merged = pd.concat(all_rows, ignore_index=True)


before = len(merged)
merged = merged.drop_duplicates(subset=["text"]).reset_index(drop=True)
after = len(merged)
print(f"Merged rows: {after} (dropped {before-after} duplicates)")


merged = merged[(merged["title"].str.len()>0) | (merged["ingredients"].str.len()>0)]
merged = merged.reset_index(drop=True)


csv_out = DATA_DIR / "recipes_merged.csv"
txt_out = DATA_DIR / "recipes_gpt2.txt"
merged.to_csv(csv_out, index=False)
with open(txt_out, "w", encoding="utf-8") as f:
    for line in merged["text"]:
        f.write(line.strip() + "\n")

print("Saved:", csv_out.name, "and", txt_out.name, "in", DATA_DIR.resolve())
print("Preview:")
display(merged.head(5)[["title","ingredients","instructions","text"]])
```

```python
import math, random, os
from pathlib import Path
import pandas as pd
import torch
from torch.utils.data import DataLoader

try:
    from transformers import (
        GPT2TokenizerFast, GPT2LMHeadModel,
        DataCollatorForLanguageModeling, get_cosine_schedule_with_warmup
    )
    from datasets import Dataset, DatasetDict
except Exception:
    !pip -q install transformers==4.44.2 datasets==3.0.1 accelerate==0.34.2 -U
    from transformers import (
        GPT2TokenizerFast, GPT2LMHeadModel,
        DataCollatorForLanguageModeling, get_cosine_schedule_with_warmup
    )
    from datasets import Dataset, DatasetDict

DATA_DIR = Path("data_recipes")
csv_path = DATA_DIR / "recipes_merged.csv"
assert csv_path.exists(), "recipes_merged.csv not found. Run Step 2 first."

df = pd.read_csv(csv_path)
df = df[df["text"].astype(str).str.contains(r"<\|recipe\|>") &
        df["text"].astype(str).str.contains(r"<\|endofrecipe\|>")].dropna(subset=["text"]).reset_index(drop=True)

print("Total recipes available:", len(df))

N_TRAIN = 20000
N_VAL   = 2000

random.seed(42)
perm = list(range(len(df)))
random.shuffle(perm)
cut = int(0.9 * len(perm))
train_idx, val_idx = perm[:cut], perm[cut:]
if N_TRAIN is not None: train_idx = train_idx[:min(N_TRAIN, len(train_idx))]
if N_VAL   is not None: val_idx   = val_idx[:min(N_VAL,   len(val_idx))]

train_texts = df.loc[train_idx, "text"].tolist()
val_texts   = df.loc[val_idx,   "text"].tolist()

raw = DatasetDict({
    "train": Dataset.from_dict({"text": train_texts}),
    "validation": Dataset.from_dict({"text": val_texts}),
})
print("Train examples:", len(raw["train"]), "| Val examples:", len(raw["validation"]))

SPECIAL_TOKENS = {"additional_special_tokens": ["<|recipe|>", "<|endofrecipe|>"]}
tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")
added = tokenizer.add_special_tokens(SPECIAL_TOKENS)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
print("Added special tokens:", added, "| pad_token:", tokenizer.pad_token)

BLOCK_SIZE = 512
def tok_fn(batch):
    return tokenizer(
        batch["text"],
        truncation=True,
        max_length=BLOCK_SIZE,
        padding=False,
        return_tensors=None,
    )

tokenized = raw.map(tok_fn, batched=True, desc="Tokenizing")
collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

tokenized.set_format(type="torch", columns=["input_ids", "attention_mask"])
train_loader = DataLoader(tokenized["train"], batch_size=2, shuffle=True, collate_fn=collator)
val_loader   = DataLoader(tokenized["validation"], batch_size=2, shuffle=False, collate_fn=collator)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = GPT2LMHeadModel.from_pretrained("gpt2")
model.resize_token_embeddings(len(tokenizer))
model.to(device)
model.train()

optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5, weight_decay=0.01)
num_epochs = 1
grad_accum = 8
num_update_steps_per_epoch = math.ceil(len(train_loader) / grad_accum)
max_train_steps = num_epochs * num_update_steps_per_epoch
scheduler = get_cosine_schedule_with_warmup(
    optimizer,
    num_warmup_steps=int(0.03 * max_train_steps),
    num_training_steps=max_train_steps
)

scaler = torch.cuda.amp.GradScaler(enabled=torch.cuda.is_available())

def evaluate():
    model.eval()
    losses = []
    with torch.no_grad():
        for batch in val_loader:
            batch = {k: v.to(device) for k, v in batch.items()}
            with torch.cuda.amp.autocast(enabled=torch.cuda.is_available()):
                outputs = model(**batch)
                loss = outputs.loss
            losses.append(loss.detach().float().cpu().item())
    model.train()
    if not losses:
        return float("inf"), float("inf")
    avg_loss = sum(losses)/len(losses)
    ppl = math.exp(avg_loss) if avg_loss < 20 else float("inf")
    return avg_loss, ppl

best_loss = float("inf")
best_dir = "gpt2_recipe_ft_custom"
os.makedirs(best_dir, exist_ok=True)

print(f"Device: {device} | GradAccum: {grad_accum} | Steps/Epoch: {num_update_steps_per_epoch}")

global_step = 0
for epoch in range(num_epochs):
    running_loss = 0.0
    optimizer.zero_grad(set_to_none=True)
    for step, batch in enumerate(train_loader):
        batch = {k: v.to(device) for k, v in batch.items()}
        with torch.cuda.amp.autocast(enabled=torch.cuda.is_available()):
            outputs = model(**batch)
            loss = outputs.loss / grad_accum
        scaler.scale(loss).backward()
        running_loss += loss.item()

        if (step + 1) % grad_accum == 0:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad(set_to_none=True)
            scheduler.step()
            global_step += 1

            if global_step % 100 == 0:
                print(f"step {global_step:>6} | train_loss {running_loss:>.4f}")
                running_loss = 0.0

    val_loss, val_ppl = evaluate()
    print(f"Epoch {epoch+1}/{num_epochs} | val_loss: {val_loss:.4f} | val_ppl: {val_ppl:.2f}")


    if val_loss < best_loss:
        best_loss = val_loss
        model.save_pretrained(best_dir)
        tokenizer.save_pretrained(best_dir)
        print("Saved new best model to:", best_dir)

print("Best validation loss:", best_loss)
```
